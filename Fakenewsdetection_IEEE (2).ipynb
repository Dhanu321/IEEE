```python
import pandas as pd

# Load ARPFN Datasets
users_df = pd.read_csv("Users.csv")
statistical_df = pd.read_csv("statistical_features.csv")
profile_df = pd.read_csv("profile_features.csv")
emotional_df = pd.read_csv("emotional_features.csv")

# Load all three Liar2 datasets
liar_train = pd.read_csv("liar2_train.csv")
liar_valid = pd.read_csv("liar2_validation.csv")
liar_test = pd.read_csv("liar2_test.csv")

# Merge all Liar2 splits
liar2_df = pd.concat([liar_train, liar_valid, liar_test], ignore_index=True)

# Merge ARPFN datasets
df_arpfn = users_df.merge(statistical_df, on="user_id").merge(profile_df, on="user_id").merge(emotional_df, on="user_id")

# Remove profile-related features (as per instructions)
profile_features = [
    'statuses_count', 'followers_count', 'following_count', 'favourites_count',
    'listed_count', 'default_profile', 'verified', 'tweet_freq',
    'follower_growth_rate', 'following_growth_rate', 'listed_growth_rate',
    'followers_following_ratio', 'screen_name_length', 'digits_in_screen_name',
    'name_length', 'digits_in_name', 'description_length'
]
df_arpfn.drop(columns=profile_features, errors='ignore', inplace=True)

# Concatenate ARPFN and Liar2 after merging all Liar2 splits
df_merged = pd.concat([
    df_arpfn.assign(source="ARPFN"),
    liar2_df[['statement', 'label']].assign(source="Liar2")  # Keep statement and label
], ignore_index=True)

# Save the final merged dataset
df_merged.to_csv("final_merged_dataset.csv", index=False)

print("âœ… Final Dataset Merged Successfully! New Shape:", df_merged.shape)
print("ðŸ”¹ Columns in Final Merged DataFrame:\n", df_merged.columns)
```

```python
import pandas as pd

# Load all three LIAR2 datasets
train_df = pd.read_csv("liar2_train.csv")
valid_df = pd.read_csv("liar2_validation.csv")
test_df = pd.read_csv("liar2_test.csv")

# Combine them into one dataset
liar_df = pd.concat([train_df, valid_df, test_df], ignore_index=True)

# Keep only the relevant columns (text and labels)
liar_df = liar_df[['statement', 'label']]

# âœ… Check the shape and sample data
print("âœ… LIAR2 Dataset Combined Successfully! Shape:", liar_df.shape)
print("\nðŸ”¹ Sample Data:\n", liar_df.head(5))
```

```python
# Load ARPFN datasets
users_df = pd.read_csv("Users.csv")
emotional_df = pd.read_csv("emotional_features.csv")
profile_df = pd.read_csv("profile_features.csv")
statistical_df = pd.read_csv("statistical_features.csv")

# Create an artificial index for merging
users_df['merge_key'] = users_df.index
liar_df['merge_key'] = liar_df.index

# Merge ARPFN (users_df) with LIAR2 (liar_df)
merged_df = users_df.merge(statistical_df, on="user_id", how="left") \
                    .merge(emotional_df, on="user_id", how="left") \
                    .merge(profile_df, on="user_id", how="left") \
                    .merge(liar_df, on="merge_key", how="left")

# Drop unnecessary columns
merged_df.drop(columns=['merge_key'], inplace=True)

# âœ… Check if 'statement' is now included
print("âœ… Final Merged Dataset Shape:", merged_df.shape)
print("\nðŸ”¹ Columns in Final Merged DataFrame:\n", merged_df.columns)

# Check if any statements are missing
missing_text = merged_df['statement'].isnull().sum()
print("\nðŸ” Missing Statements:", missing_text)

# Show sample merged data
print("\nðŸ”¹ Sample Merged Data:\n", merged_df[['user_id', 'statement', 'label']].head(10))
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Initialize stopwords and stemmer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    """Cleans and preprocesses text data."""
    if pd.isna(text):
        return ""  # Handle NaN values

    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation and numbers
    words = word_tokenize(text)  # Tokenize
    words = [stemmer.stem(word) for word in words if word not in stop_words]  # Remove stopwords & stem
    
    return " ".join(words)

# âœ… Apply preprocessing on the `statement` column
merged_df['clean_statement'] = merged_df['statement'].apply(preprocess_text)

# âœ… Check Sample Processed Data
print("âœ… Sample Processed Statements:\n", merged_df[['statement', 'clean_statement']].head())
```
